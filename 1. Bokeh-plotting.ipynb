```python
p = figure(plot_width= 400, plot_height = 400 )

p.circle([1,2,3,4,5],[6,7,2,4,5], size = 15, line_color='navy', fill_color='orange', fill_alpha = 0.5)
show(p)
```

```python
# square instaed of a circle

p = figure(plot_width = 400, plot_height = 400)
p.square([1,2,3,4,5],[6,7,2,4,5],color='firebrick', size=[10,15,20,12,25], alpha=0.6)
show(p)
```

## Basic Line plot

```python
p = figure(plot_width = 400, plot_height = 400)
p.line([1,2,3,4,5],[6,7,2,4,5],line_width=2)
show(p)
```

## Images

```python
import numpy as np
N = 20
img = np.empty((N,N), dtype=np.uint32)
view = img.view(dtype=np.uint8).reshape(N,N,4)

for i in range(N):
    for j in range(N):
        view[i, j, 0 ] = int(i/N*235)
        view[i, j, 1 ] = 55
        view[i, j, 2 ] = int(j/N*235)
        view[i, j, 3 ] = 255
p = figure(x_range=[0,10], y_range=[0,10])

p.image_rgba(image=[img], x=[0], y=[0], dw=[10], dh=[10])
show(p)
```

```python
x = [1,2,3,4,5]
y = [6, 7, 8, 7, 3]

p = figure(plot_width=400, plot_height= 400)
p.line(x, y, line_width = 2)
p.circle(x, y, fill_color = 'firebrick', size = 8)

show(p)
```
